```python
import pandas as pd
import numpy as np
import re
import datetime as dt

# Se inician variables para mantener un formato consistentes en los menus
anchoConsola = 80
delimitador = "\n" + "-" * anchoConsola + "\n"
textoMenuSecciones = "En esta seccion usted puede: "
textoVolverMenu = "> Volviendo al menu anterior <".center(anchoConsola, "-")
textoParaVolver = "Si desea volver escriba \"Volver\""


# Se crean funciones para guardar los archivos, son 3 tipos de guardados, 2 para los acrhivos maestros y 1 general para los detalles y resumenens.
def GuardarArchivoProductos():
    print("".center(anchoConsola, "-"))
    nombreArchivo = "archivoMaestroProductos"
    
    try:
        archivoMaestroProductos.to_csv(f"{nombreArchivo}.csv")
        print(f"Archivo con nombre {nombreArchivo}.csv se guardo con exito")
    except:
        print(f"Archivo con nombre {nombreArchivo}.csv no se logro guardar")


def GuardarArchivoFacturas():
    print("".center(anchoConsola, "-"))
    nombreArchivo = "archivoMaestroFacturas"
    
    try:
        archivoMaestroFacturas.to_csv(f"{nombreArchivo}.csv")
        print(f"Archivo con nombre {nombreArchivo}.csv se guardo con exito")
    except:
        print(f"Archivo con nombre {nombreArchivo}.csv no se logro guardar")   


def GuardarArchivo(resumen):
    nombreArchivo = input("Ingrese el nombre del archivo, entre 2 a 10 caracteres, con almenos 1 letra:\n")
    nombreArchivo = ValidarTexto(nombreArchivo, 2)
    
    try:
        resumen.to_csv(f"{nombreArchivo}.csv")
        print(f"Archivo con nombre {nombreArchivo}.csv se guardo con exito")
    except:
        print(f"Archivo con nombre {nombreArchivo}.csv no se logro guardar")
    pass


# Se crea funcion para abrir los archivos maestros
def AbrirArchivo(nombreArchivo):
    try:
        archivoDF = pd.read_csv(f"{nombreArchivo}.csv", index_col=0)
        print(f"{nombreArchivo}.csv cargado con exito <".ljust(anchoConsola, "-"))
        return archivoDF
    except:
        return False

# Se crean 3 funciones para validar los datos y asegurar que el input no haga fallar el programa
def ValidarDigito(numero):
#     Se valida si el numero ingresado es un entero, en caso contrario solicita nuevo ingreso
    while True:
        if numero.isdigit():
            return int(numero)
        elif numero.lower() == "volver":
            print(textoVolverMenu)
            return numero.lower()
        else:
            numero = input("Valor invalido, ingrese un numero entero:\n")
            pass
        
def ValidarTexto(texto, cotaInferior= 2, cotaSuperior= 20, listaObjetoUnico = []):
#     Se valida que se ingrese un texto, que no salga de las cotas y si es requerido que no este en la lista.
    while True:
        if texto.isdigit():
            texto = input("El texto debe contener almenos una letra, intente nuevamente:\n")
            pass
        elif len(texto) <= cotaInferior or len(texto) >= cotaSuperior:
            texto = input(f"Debe ingresar entre {cotaInferior} y {cotaSuperior} caracteres:\n")
            pass
        elif texto.lower() == "volver":
            print(textoVolverMenu)
            return texto.lower()
        elif texto.lower() in listaObjetoUnico:
            texto = input(f"Nombre ya ingresado, porfavor ingrese nombres unicos: \n")
            pass
        else:
            return texto.lower()

def ValidarOpciones(opcionesDisponibles):
#     Se valida que el valor ingresado se encuentre dentro de las opciones especificadas.
    opcionIngresada = input("Ingrese numero de la opcion que quiere realizar:\n")
    
    while not opcionIngresada in opcionesDisponibles.keys():
        opcionIngresada = input("Opcion incorrecta, intentelo nuevamente:\n")
        
    return opcionIngresada
```

a) Ingresar producto: debe ingresarse el nombre y el precio del producto. Tendrá un sub menú: nuevo producto, guardar archivo o volver.

```python
"""
a) Modulo contiene 2 funciones, MenuIngresoProducto es el menu principal que llama a NuevoProducto para ir agregando datos al DF
"""

def NuevoProducto(copiaDF):
#     Se genera loop para agregar productos en continuo
    seccion = "> Ingreso de nuevo producto <"
    while True: 
        print(seccion.center(anchoConsola, "="))
        print(textoParaVolver.center(anchoConsola))
        nombreProducto = input("Ingrese Nombre del Producto entre 2 a 20 caracteres, con almenos 1 letra:\n")
        nombreProducto = ValidarTexto(nombreProducto, listaObjetoUnico = copiaDF["Nombre Producto"].values)
        
        if nombreProducto == "volver":
            break

        precioProducto = input("Ingrese Precio del Producto como numero entero:\n")
        precioProducto = ValidarDigito(precioProducto)
        
        if precioProducto == "volver":
            break
#         Se agrega el producto y precio al final de la copia del dataframe, para posteriormente descartar o agregar los datos al archivo maestro
        copiaDF.loc[len(copiaDF)] = {"Nombre Producto": nombreProducto, "Precio": precioProducto}
        print("Producto ingresado correctamente".center(anchoConsola, "_"))
        pass

    return copiaDF

def MenuIngresoProducto():
    global archivoMaestroProductos
    copiaDF = archivoMaestroProductos.copy()
    
    menu = "> Ingreso de Productos <"
    opcionesDisponibles = {"1": "Ingreso de nuevo producto", "2": "Guardar cambios e ingresar mas productos", 
                           "8": "Guardarc cambios y volver", "9": "volver sin guardar"}
    
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        
        if not copiaDF.empty:
            print(copiaDF, end=delimitador)
        
        print(textoMenuSecciones)
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")

        opcionElegida = ValidarOpciones(opcionesDisponibles)

        if opcionElegida == "1":
#             Se deriva a sub-funcion y se retorna DF modificado
            copiaDF = NuevoProducto(copiaDF)
        elif opcionElegida == "2":
#             Se modifica el archivomaestro en ram y se usa funcion para guardar en disco duro
            archivoMaestroProductos = copiaDF
            GuardarArchivoProductos()
        elif opcionElegida == "8":
#             Igual a opcion 2 pero se rompe el loop del menu
            archivoMaestroProductos = copiaDF
            GuardarArchivoProductos()
            print(textoVolverMenu)
            break
        elif opcionElegida == "9":
#             Se rompe loop del menu
            print(textoVolverMenu)
            break


# MenuIngresoProducto()
# print(archivoMaestroProductos)
# precioProducto = archivoMaestroProductos[archivoMaestroProductos["Nombre Producto"] == "caca"]["Precio"].iloc[0]
# print(precioProducto)
```

b) Ingresar factura: debe ingresarse el nombre del cliente, luego deberá generar un submenú: ingresar producto, guardar y salir.

Los productos se ingresan con nombre y precio. La factura además debe guardar la fecha y hora actual de la venta.

```python
"""
b) Modulo contiene 3 funciones, 
MenuIngresarFactura es el menu principal de entrada que deriva a CreacionDeFactura.
CreacionDeFactura es submenu para agregar facturas en continuo junto a los productos, este llama a GuardadoDeFactura.
GuardadoDeFactura se encarga de guardar la factura en la base de datos.
"""

def GuardadoDeFactura(numeroFactura, nombreCliente, productosEnFactura):
#     Como esta funcion cambia el df principal se llama desde la funcion.
    global archivoMaestroFacturas
#     Se genera un DF para iterar agregando la lista productosEnFactura con los datos del cliente y factura correspondiente.
    productosDF = pd.DataFrame(columns=["Numero de Factura", "Nombre Cliente", "Producto", "Precio", "Cantidad", "Fecha Ingreso", "Subtotal"])
    for producto, precio, cantidad in productosEnFactura:
        productosDF.loc[len(productosDF)] = {"Numero de Factura": numeroFactura, "Nombre Cliente": nombreCliente,
                                               "Producto": producto,
                                               "Precio": precio,
                                               "Cantidad": cantidad,
                                               "Fecha Ingreso": dt.datetime.now().strftime("%d-%m-%Y ; %H:%M:%S"),
                                                "Subtotal": precio * cantidad}
    
#     Se concatena el DF generado con el maestro de facturas y el resultado se alimenta directo a la memoria del maestro factura
    archivoMaestroFacturas =  pd.concat([archivoMaestroFacturas, productosDF], ignore_index=True)
    GuardarArchivoFacturas()
    print(f"Factura {numeroFactura} del cliente {nombreCliente} guardada con exito ")
    return
    

def CreacionDeFactura(nombreCliente):
    global archivoMaestroProductos
    seccion = "> Ingresar nueva factura <"
    opcionesDisponibles = {"1": "Agregar Producto", "2": "Guardar y agregar nueva factura", 
                           "8": "Guardar Factura y volver", "9": "Volver sin guardar"}   
#     Se genera lista que se alimentara con los productos a agregar en la factura
    productosEnFactura = []
            
    while True:
#         Se da valor a la factura segun valor de la anterior
        if not archivoMaestroFacturas.empty:
            numeroFactura = archivoMaestroFacturas["Numero de Factura"].max() + 1
        else:
            numeroFactura = 1
        
        print(seccion.center(anchoConsola, "="))
#         Se imprime datos de la factura a ingresar
        print(f"Factura {numeroFactura} >".ljust(anchoConsola//2,"-")+ f"< Cliente {nombreCliente}".rjust(anchoConsola//2,"-"))
    
        print(f"producto".ljust(anchoConsola//3,"-")+ f"precio".rjust(anchoConsola//3,"-") + f"cantidad".rjust(anchoConsola//3,"-"))
        
        for producto, precio, cantidad in productosEnFactura:
            print(f"{producto}".ljust(anchoConsola//3,"-")+ f"{precio}".rjust(anchoConsola//3,"-") +f"{cantidad}".rjust(anchoConsola//3,"-"))
        print(delimitador, end="")

        print(textoMenuSecciones)
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
            
        opcionElegida = ValidarOpciones(opcionesDisponibles)

        if opcionElegida == "1":
#             Se procede a tomar los datos de producto, si no estan en el maestro de productos se ingresa como producto nuevo
            nombreProducto = input("Ingrese Nombre del Producto entre 2 a 20 caracteres, con almenos 1 letra:\n")
            nombreProducto = ValidarTexto(nombreProducto)
            
            if not nombreProducto in archivoMaestroProductos["Nombre Producto"].values:
                precioProducto = input("Ingrese Precio del Producto como numero entero:\n")
                precioProducto = ValidarDigito(precioProducto)
                archivoMaestroProductos.loc[len(archivoMaestroProductos)] = {"Nombre Producto": nombreProducto, "Precio": precioProducto}
                GuardarArchivoProductos()
            else:
                precioProducto = archivoMaestroProductos[archivoMaestroProductos["Nombre Producto"] == nombreProducto]["Precio"].iloc[0]
            
            cantidadProducto = input("Ingrese cantidad del Producto como numero entero:\n")
            cantidadProducto = ValidarDigito(cantidadProducto)
#             Se agregan datos del producto a lista de productos en factura
            productosEnFactura.append([nombreProducto, precioProducto, cantidadProducto])
        
        
        elif opcionElegida == "2": 
#             Se agregan los datos a la base de datos de la factura
            GuardadoDeFactura(numeroFactura, nombreCliente, productosEnFactura)
#             Se reinicia lista de productos y se solicita nombre de cliente nuevo
            productosEnFactura = []
            nombreCliente = input("Ingrese el nombre del cliente, entre 2 a 20 caracteres, con almenos 1 letra:\n")
            nombreCliente = ValidarTexto(nombreCliente)
            
        elif opcionElegida == "8":
#             Se agregan los datos a la base de datos de la factura y se sale del loop
            GuardadoDeFactura(numeroFactura, nombreCliente, productosEnFactura)
    
            print(textoVolverMenu)
            break
            
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break

def MenuIngresarFactura():
    menu = "> Ingreso de Factura <"
    opcionesDisponibles = {"1": "Ingresar Nueva factura", "9": "volver"}
        
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        
        print(textoMenuSecciones)
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")

        opcionElegida = ValidarOpciones(opcionesDisponibles)

        if opcionElegida == "1":
#             Se ingresa a subFuncion para ingresar facturas
            nombreCliente = input("Ingrese el nombre del cliente, entre 2 a 20 caracteres, con almenos 1 letra:\n")
            nombreCliente = ValidarTexto(nombreCliente)
            CreacionDeFactura(nombreCliente)
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break

    pass


# CreacionDeFactura("riqui")
# print(archivoMaestroFacturas)
# MenuIngresarFactura()
```

```python
"""
c) Modulo contiene 1 funcion encargada de mostrar el menu y desplegar el resumen, para guardar llama a funcion GuardarArchivo de primera celda.
"""

def MenuVerResumen():
    menu = "> Visualizar Resumen <"
    opcionesDisponibles = {"1": "Guardar Resumen y volver", "9": "Volver"}
    
    print(delimitador, end="")
    print(menu.center(anchoConsola, "="), end = delimitador)
#     Se desplega ek resumen agrupando las facturas y sumando los subtotales, se cambia nombre de columna para que se adecue al contexto
    resumen = archivoMaestroFacturas[["Numero de Factura","Nombre Cliente", "Subtotal"]].groupby(by = ["Numero de Factura","Nombre Cliente"]).sum()
    resumen.rename(columns = {"Subtotal": "Total vendido"}, inplace =True)
    
    print(resumen, end=delimitador)

    print(textoMenuSecciones)
    for opcion in opcionesDisponibles.keys():
        print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
    opcionElegida = ValidarOpciones(opcionesDisponibles)      
    
    if opcionElegida == "1":
        GuardarArchivo(resumen)
        MenuVerResumen()
    elif opcionElegida == "9":
        print(textoVolverMenu)
# MenuVerResumen()
# print(archivoMaestroFacturas.columns)
```

d) Ver detalle: se debe tener las opciones desplegar: 
    -ventas por usuario, 
    - ventas por producto, 
    - ventas diarias, 
    - guardar vista (ingresando el nombre de archivo).

```python
"""
d) Modulo contiene 4 funciones, 
MenuVerDetalle es el menu principal de entrada que se encarga de desplegar los detalles y llamar a funciones creadoras de las tablas.
VentasXCliente agrupa el maestro de facturas segun cliente y valor total de sus compras
VentasXProducto agrupa el maestro de facturas segun producto y valor total comprado
VentasXDia agrupa el maestro de facturas segun dia y valor total comprado por dia, creando indices indicador de mes y año para agrupar los dias.
"""


def VentasXCliente():
    ventasCliente = archivoMaestroFacturas[["Nombre Cliente", "Subtotal"]].groupby(by = ["Nombre Cliente"]).sum()
    ventasCliente.rename(columns = {"Subtotal": "Total vendido"}, inplace =True)
    return ventasCliente

def VentasXProducto():
    ventasProducto = archivoMaestroFacturas[["Producto", "Cantidad", "Subtotal"]].groupby(by = ["Producto"]).sum()
    ventasProducto.rename(columns = {"Subtotal": "Total vendido"}, inplace =True)
    return ventasProducto

def VentasXDia():
    ventasDia = archivoMaestroFacturas[["Fecha Ingreso", "Subtotal"]].copy()
    ventasDia["Fecha Ingreso"] = pd.to_datetime(ventasDia["Fecha Ingreso"], dayfirst=True)
    ventasDia["Año"] = ventasDia["Fecha Ingreso"].dt.year
    ventasDia["Mes"] = ventasDia["Fecha Ingreso"].dt.month
    ventasDia["Dia"] = ventasDia["Fecha Ingreso"].dt.day
    ventasDia = ventasDia[["Año", "Mes", "Dia", "Subtotal"]].groupby(["Año", "Mes", "Dia"]).sum()
    ventasDia.rename(columns = {"Subtotal": "Total vendido"}, inplace =True)
    return ventasDia

def MenuVerDetalle():
    menu = "> Visualizar Detalles <"
    opcionesDisponibles = {"1": "Ver ventas por cliente", "2": "Ver ventas por producto",
                           "3": "Ver ventas diarias", "4": "Guardar detalle", "9": "Volver"}
#     Estas variables se utilizaran como posicionadores del detalle a desplegar
    contexto = ""
    TituloDetalle = ""
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        
#        Se detecta si contexto es un DF, si es DF imprimira el detalle y su titulo
        if isinstance(contexto, pd.DataFrame):
            print(TituloDetalle.center(anchoConsola, "-"))
            print(contexto, end=delimitador)
        
        print(textoMenuSecciones)
        
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
        opcionElegida = ValidarOpciones(opcionesDisponibles)      

        if opcionElegida == "1":
            contexto = VentasXCliente()
            TituloDetalle = "> Ventas por Cliente <"
        if opcionElegida == "2":
            contexto = VentasXProducto()
            TituloDetalle = "> Ventas por Producto <"
        if opcionElegida == "3":
            contexto = VentasXDia()
            TituloDetalle = "> Ventas por Dia <"
        if opcionElegida == "4":
            GuardarArchivo(contexto)
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break

            
            
# print(archivoMaestroFacturas.columns)
# MenuVerDetalle()
```

e) Borrar datos: elimina los archivos guardados (elegir productos, facturas, todos o volver)

```python
"""
e) Modulo contiene 5 funciones, 
MenuBorrarDatos es el menu principal de entrada que se encarga de direccionar el tipo de borrado que se desea aplicar
AdvertenciaAlBorrar despliega una advertencia para confirmar que deseas borrar los datos
BorrarProducto sub-menu encargado de borrar los productos segun nombre ingresado en BD de productos y facturas.
BorrarFactura sub-menu encargado de borrar las facturas segun numero ingresado en BD de facturas.
BorrarTodosLosArchivos funcion que despliega una segunda advertencia y dado la confirmacion borra todos los datos guardados dejando los archivos vacios
"""

def BorrarProducto():
    global archivoMaestroProductos
    global archivoMaestroFacturas

    menu = "> Borrar productos guardados <"
    opcionesDisponibles = {"1": "Ingresar nombre de producto a borrar", "8": "Guardar cambios y volver", 
                           "9": "Volver sin guardar cambios"}
    
    copiaDFProductos = archivoMaestroProductos.copy()
    copiaDFFacturas = archivoMaestroFacturas.copy()
    
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        print("Los produtos Guardados son: ")
        
        print(copiaDFProductos, end=delimitador)

        print(textoMenuSecciones)
        
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
        opcionElegida = ValidarOpciones(opcionesDisponibles)
        
        if opcionElegida == "1":
#             Se solicita nombre del producto, si esta en la BD se procede a borrar, si no se solicita otro ingreso
            productoEliminar = input("Ingrese Producto=>")
            productoEliminar = ValidarTexto(productoEliminar)
            
            while not productoEliminar.lower() in copiaDFProductos["Nombre Producto"].values:
                print("Producto no encontrado en la base de datos, intente nuevamente")
                productoEliminar = input("Ingrese Producto Nuevamente =>")
                productoEliminar = ValidarTexto(productoEliminar)
                
#             Se elimina las filas del match con funcion drop alimentando los indeces donde hubo match
            copiaDFProductos = copiaDFProductos.drop(copiaDFProductos[copiaDFProductos["Nombre Producto"] == productoEliminar].index)
            try:
#                 Se repite formula pero para borrar BD de facturas si es que hay coincidencia
                copiaDFFacturas = copiaDFFacturas.drop(copiaDFFacturas[copiaDFFacturas["Producto"] == productoEliminar].index)
            except:
                pass

        elif opcionElegida == "8":
#             se guarda los BD copiados en los BD originales para luego guardar los cambios en los archivos
            archivoMaestroProductos = copiaDFProductos
            GuardarArchivoProductos()
            archivoMaestroFacturas = copiaDFFacturas
            GuardarArchivoFacturas()
            break
            
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break
        else:
            pass

def BorrarFactura():
    global archivoMaestroFacturas
    menu = "> Borrar facturas guardados <"
    opcionesDisponibles = {"1": "Ingresar numero de factura a borrar", "8": "Guardar cambios y volver", 
                           "9": "Volver sin guardar cambios"}
    
    copiaDFFacturas = archivoMaestroFacturas.copy()
    
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        print("Las facturas guardados son: ")
        
        print(copiaDFFacturas.to_markdown(), end=delimitador)

        print(textoMenuSecciones)
        
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
        opcionElegida = ValidarOpciones(opcionesDisponibles)
        
        if opcionElegida == "1":
#             Se solicita numero de la factura, si esta en la BD se procede a borrar, si no se solicita otro ingreso
            numeroFactura = input("Ingrese Factura =>")
            numeroFactura = ValidarDigito(numeroFactura)
            
            while not numeroFactura in copiaDFFacturas["Numero de Factura"].values:
                print("Factura no encontrado en la base de datos, intente nuevamente")
                numeroFactura = input("Ingrese Factura nuevamente =>")
                numeroFactura = ValidarDigito(numeroFactura)
#             Se elimina las filas del match con funcion drop alimentando los indeces donde hubo match
            copiaDFFacturas = copiaDFFacturas.drop(copiaDFFacturas[copiaDFFacturas["Numero de Factura"] == numeroFactura].index)
            
            pass

        elif opcionElegida == "8":
#             se guardan los datos cambiados
            archivoMaestroFacturas = copiaDFFacturas
            GuardarArchivoFacturas()
            break
            
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break
        else:
            pass

def BorrarTodosLosArchivos():
    global archivoMaestroProductos
    global archivoMaestroFacturas
#     Se solicita segunda confirmacion para forrar los datos
    if AdvertenciaAlBorrar("Borrar todo los datos y los archivos guardados.\nEsto restaura la memoria ram y disco duro, esto sera irreversible."):
        
#         Se restablecen las tablas y se guarda en los archivos
        archivoMaestroProductos = pd.DataFrame(columns=["Nombre Producto", "Precio"])
        GuardarArchivoProductos()
        archivoMaestroFacturas = pd.DataFrame(columns=["Numero de Factura", "Nombre Cliente", 
                                               "Producto", "Precio", "Cantidad", "Fecha Ingreso", "Subtotal"])
        GuardarArchivoFacturas()
        print("> Archivos maestros borrados exitosamente <".center(anchoConsola, "="),end=delimitador)
    pass

def AdvertenciaAlBorrar(texto):
#     Simple loop de confirmacion y print de mensaje continuo hasta aceptar o rechazar.
    while True:
        print("> A D V E R T E N C I A <".center(anchoConsola, "%"))
        print("> El contenido eliminado no se puede restaurar <".center(anchoConsola, "-"))
        print(f"->Para continuar con {texto} ingrese \"1\"\n->De lo contrario ingrese \"9\"")
        confirmacion = input("=>")
        if confirmacion == "1":
            print("> Procediendo <".center(anchoConsola, "-"))
            return True
        elif confirmacion == "9":
            print("> Retornando <".center(anchoConsola, "-"))
            return False
        else:
            print("Fallo confirmacion, intente nuevamente")
            pass

def MenuBorrarDatos():
    menu = "> Borrar datos guardados <"
    opcionesDisponibles = {"1": "Eliminar producto segun nombre", "2": "Eliminar factura segun numero", "7": "Eliminar todo incluso los archivos", "9": "Volver"}
    
    while True:
        print(delimitador, end="")
        print(menu.center(anchoConsola, "="), end = delimitador)
        
        print(textoMenuSecciones)
        
        for opcion in opcionesDisponibles.keys():
            print(f"\t{opcion}. {opcionesDisponibles[opcion]}")
        opcionElegida = ValidarOpciones(opcionesDisponibles)      

        if opcionElegida == "1":
            if AdvertenciaAlBorrar(opcionesDisponibles["1"]):
                BorrarProducto()
            else:
                pass
        elif opcionElegida == "2":
            if AdvertenciaAlBorrar(opcionesDisponibles["2"]):
                BorrarFactura()
            else:
                pass
        elif opcionElegida == "7":
            if AdvertenciaAlBorrar(opcionesDisponibles["7"]):
                BorrarTodosLosArchivos()
                pass
            else:
                pass
        elif opcionElegida == "9":
            print(textoVolverMenu)
            break
        else:
            pass

# MenuBorrarDatos()
```
