### 1 Crear un sistema en el que un usuario pueda ingresar una cifra especificando uno de los tres distintos sistemas numéricos:
• Decimal

• Binario

• Hexadecimal

El programa debe darle al usuario un menú que indique qué tipo de número quiere ingresar, luego pedirle el número y luego debe indicar que tipo de salida requiere. Esto se debe hacer permanentemente hasta que el usuario ingrese la opción salir.

Debe construir una función para cada transformación, sin usar las funciones de transformación numérica de Python (hex, bin,int.)

```python
# Se importa Regular expresions para utilizarlos en las funciones validadoras.
import re
# Se importa Math para facilitar calculos de logaritmos en las tranformaciones
import math as m

# Se crea funciones de transformacion a parte para mantener orden.
def DecimalABinario(valor):
    valor = int(valor)
    if valor == 0:
        return "0"
#     Se tomara aproximacion sustractiva iterativa, encontrando el mayor valor de exponente 2 que pueda sustraerse al valor
#     Para esto, logaritmos de base 2 es el metodo matematico ideal.
#     Se crea una lista de exponentes para ir agregandolos y luego construir el valor binario
    exponenteLista = []
    
    while valor > 0:
#         Primero encontramos el exp en base 2 que da como resultado el valor y lo redondeamos hacia abajo.
        expBase2 = int(m.log2(valor))
#         Luego agregamos ese resultado a la lista de exponentes y le restamos el 2^exponente al valor original
        exponenteLista.append(expBase2)
        valor -= 2 ** expBase2
        
#     Luego de tenerlo iterado debemos construir el numero en binario, tomando las posiciones de los exponentes donde habra unos y el resto seria 0
#     Iniciamos una lista que contendra el resultado, la populamos con 0 y luego remplazamos con 1 en las posiciones correspondientes
    binarioList = []
    for exponenteMaximo in range(exponenteLista[0]+1):
        binarioList.append(0)
    for exponentePosicion in exponenteLista:
        binarioList.pop(exponentePosicion)
        binarioList.insert(exponentePosicion, 1)
    
#     Finalmente volteamos la lista y la transformamos en string, luego la devolvemos.
    binarioList.reverse()
    
    binarioStr = ""
    for num in binarioList:
        binarioStr += str(num)
    
    return binarioStr

# Desde decimal a hex se tomo la aproximacion de dividir en 16 de manera iterativa.
def DecimalAHexadecimal(valor):
#     Iniciamos un diccionario para dar formato hex mas adelante e iniciamos una lista para guardar los constructores del hex
    HexSobre9Dict = {10 : "a", 11: "b", 12: "c", 13: "d", 14: "e", 15: "f"}
    posicionalLista = []
    valor = int(valor)
#     Validamos el valor e iteramos en caso correspondiente
    if valor == 0:
        return "0"
    
    while valor > 0:
#         Dividiendo el valor constantemente por 16 nos dara el resto como el valor en cada posicion dependiendo del N de iteracion
        resto = valor % 16
        valor = valor // 16
#         Guardamos el resto y seguimos iterando, el arreglo de restos sera finalmente el numero hex a retornar
        posicionalLista.append(resto)
    
#     Volteamos la lista y generamos el string a devolver, tomando en consideracion si el resto es mayor a 9 darle el formato hex correspondiente
    posicionalLista.reverse()
    hexstring = ""
    for val in posicionalLista:
        if val > 9:
            hexstring += HexSobre9Dict.get(val)
        else:
            hexstring += str(val)
    return hexstring

# De bin a Dec debemos considerar el metodo f(x) = sum(i*2^n) donde i seria el valor en la posicion n del binario
def BinarioADecimal(valor):
#     Primero validamos que el valor sea valido, inicializamos el decimal a iterar y variable n, iteramos segun funcion sumatoria
    if len(valor) == 1:
        return valor
    n = 0
    Decimal = 0
#     Es requerimiento dar vuelta el decimal para iterar de manera correcta el numero binario
    for i in valor[::-1]:
        Decimal += int(i) * (2 ** n)
        n += 1
    return Decimal

# De Hec a Dec se utiliza misma formula que en el anterior f(x) = sum(i*16^n)
def HexaADecimal(valor):
#    Se inicia el diccionario para dar formato posterior al hex, se valida dato de entrada y luego se itera segun formula
    HexSobre9Dict = {"a": 10, "b": 11, "c": 12, "d": 13, "e": 14, "f": 15}
    if len(valor) == 1:
        if re.search(r"[a-fA-F]", valor):
            return HexSobre9Dict.get(valor.lower())
        else:
            return valor
    n = 0
    Decimal = 0
#     Muy similar a bin-dec, pero requiere ajustar valores por diferencia de alcance en los valores
    for i in valor[::-1]:
        if re.search(r"[a-fA-F]", i):
            Decimal += HexSobre9Dict.get(i.lower()) * (16 ** n)
        else:
            Decimal += int(i) * (16 ** n)
        n += 1
    return Decimal
    
# Se crea funcion encargada de seleccionar funcion para conversion de valor numerico
def DeterminarConvertidor(numIn, opcionIn, opcionOut):
#     Solo difiere a la funcion correcta dependiendo del input
    if opcionIn == "d":
        if opcionOut == "b":
            return DecimalABinario(numIn)
        elif opcionOut == "h":
            return DecimalAHexadecimal(numIn)
    elif opcionIn == "b":
        if opcionOut == "d":
            return BinarioADecimal(numIn)
        elif opcionOut == "h":
#             Para simplificarme la vida anide las funciones para no tener que crear nuevas funciones desde bin a hex y viceversa
            return DecimalAHexadecimal(BinarioADecimal(numIn))
    elif opcionIn == "h":
        if opcionOut == "d":
            return HexaADecimal(numIn)
        elif opcionOut == "b":
            return DecimalABinario(HexaADecimal(numIn))
    

# Se crea validador de opciones para asegurar correcto ingreso de la opcion y facilitar formateo del resto de las funciones
def ValidarOpcion(opcion, otraopcion = ""):
    dictSeleccion = {"d": "Decimal", "b": "Binario", "h": "Hexadecimal"}
    valOp = opcion
    while True:
        if valOp.isdigit():
            valOp = input("Ingreso un numero, intentelo otra vez: \n")
            pass
        elif valOp.lower() == otraopcion:
            valOp = input("Igreso la misma unidad de entrada, intentelo otra vez: \n")

        elif valOp.lower() in dictSeleccion.keys():
            
            print(f"Eligio {dictSeleccion.get(valOp.lower())}, prosiguiendo.\n")
            return valOp.lower()
        else:
            valOp = input("Opcion invalida, intentelo otra vez: ")
            pass

# Se crea validador de valor para asegurar correcto ingreso del valor para cambio y facilitar formateo del resto de las funciones
def ValidadorDeNumero(valor, opcion):
    valEvaluar = valor
    if opcion == "d":
        while True:
            if valEvaluar.isdigit():
                print("Valor es digito, prosiguiendo.\n")
                return valEvaluar
            else:
                valEvaluar = input("Valor no es digito, ingrese otra vez: ")
                pass
    elif opcion == "b":
        while True:
#             Utilizando RE se evalua si valor corresponde a binario o a hex
            if not re.search(r"[^01]", valEvaluar):
                print("Valor es binario, prosiguiendo.\n")
                return valEvaluar
            else:
                valEvaluar = input("Valor no es binario, ingrese otra vez: ")
                pass
    elif opcion == "h":
        while True:
            if not re.search(r"[^a-fA-F0-9]", valEvaluar):
                print("Valor es hexadecimal, prosiguiendo.\n")
                return valEvaluar
            else:
                valEvaluar = input("Valor no es hexadecimal, ingrese otra vez: ")
            
# Esta es la funcion que coordina los trigers del resto de funciones, dandole un orden a la syntaxis.
def ProgramaConvertidor():
    print("¡¡Bienvenido al convertidor de numero!!\n Para salir ingrese \"salir\"")

    dictSeleccion = {"d": "Decimal", "b": "Binario", "h": "Hexadecimal"}
    while True:
    #     1° solicita el tipo de valor de entrada
        print("""
De la siguiente lista seleccione que tipo de numero ingresara:\n
    - ingrese "d" para Decimal
    - ingrese "b" para Binario
    - ingrese "h" para Hexadecimal
        """)


        opcionIn = input("Opcion:\n")
        if re.search("salir", opcionIn):
                print("Saliendo" + "_" * 50)
                break
        opcionIn = ValidarOpcion(opcionIn)

    #     2° se solicita el valor de entrada
        numIn = input("Ingrese numero a convertir: \n")
        if re.search("salir", numIn):
                print("Saliendo" + "_" * 75)
                break
        numIn = ValidadorDeNumero(numIn, opcionIn)

    #     3° se solicita el tipo de valor de salida
        print("""
De la siguiente lista seleccione a que tipo de numero desea convertir:\n
    - ingrese "d" para Decimal
    - ingrese "b" para Binario
    - ingrese "h" para Hexadecimal
        """)
        opcionOut = input("Opcion: \n")
        if re.search("salir", opcionOut):
                print("Saliendo" + "_" * 70)
                break
        opcionOut = ValidarOpcion(opcionOut, opcionIn)

    #     4° se procede a generar el valor de salida segun inputs ya validadas 
        numout = DeterminarConvertidor(numIn, opcionIn, opcionOut)

        print(f"Al transformar el valor {numIn} que es {dictSeleccion.get(opcionIn.lower())} a {dictSeleccion.get(opcionOut.lower())} resulta {numout}")
        print("_" * 60)
    
    print("Has salido del programa, muchas gracias por utilizar este programa!!!")


# Se ejecuta la funcion principal
ProgramaConvertidor()
```

### 2. Crear un sistema que almacene una lista de compras. Para cada ítem, debe almacenar el código (ej. de barras). El nombre del producto, el precio unitario y la cantidad comprada. Al finalizar el ingreso, se deberá emitir un informe con:

Nombre del producto, cantidad, subtotal

-------------------------------------------------
IVA y total de la compra.

```python
# Se copio codigo para transformar de decimal a binario y viceversa del ejercicio anterior, 
# Se borro los comentarios para evitar redundancias
def DecimalABinario(valor):
    valor = int(valor)
    if valor == 0:
        return "0"
    exponenteLista = []
    
    while valor > 0:
        expBase2 = int(m.log2(valor))
        exponenteLista.append(expBase2)
        valor -= 2 ** expBase2
        
    binarioList = []
    for exponenteMaximo in range(exponenteLista[0]+1):
        binarioList.append(0)
    for exponentePosicion in exponenteLista:
        binarioList.pop(exponentePosicion)
        binarioList.insert(exponentePosicion, 1)
    
    binarioList.reverse()
    
    binarioStr = ""
    for num in binarioList:
        binarioStr += str(num)
    
    return binarioStr

def BinarioADecimal(valor):
    if len(valor) == 1:
        return valor
    n = 0
    Decimal = 0
    for i in valor[::-1]:
        Decimal += int(i) * (2 ** n)
        n += 1
    return Decimal



# Funcion para validar que se ingresan numeros
def ValidarDigito(numero):
    while True:
        if numero.isdigit():
            return numero
        else:
            numero = input("Valor invalido, ingrese un numero: ")
            pass

# Utilizando codigo binario se crea codigo de producto correlativo al numero de ingreso
def GeneradorCodigo(masterData):
    if len(masterData) == 0:
        return "0"
    
    codUltimoIn = masterData[-1][0]
    codOut = DecimalABinario(int(BinarioADecimal(codUltimoIn)) + 1)
    return codOut

# Funcion para agrandar codigos pequeños
def FormatearCodigo(codigo):
    largoCod = len(codigo)
    if largoCod <= 5:
        codigoOut = "0" * (5 - largoCod)
        return codigoOut + codigo
    else:
        return codigo

# Se crea funcion para dar formato y obtener un resultado mas uniforme en la boleta
def FormatearItem(item, anchoFormato, divisiones = 4, alineacion = "L", espaciador = " "):
    if alineacion == "L":
        itemFormateado = item + (espaciador * int(anchoFormato / divisiones - len(item)))
    elif alineacion == "R":
        itemFormateado = (espaciador * int(anchoFormato / divisiones - len(item))) + item
    elif alineacion == "C":
        itemFormateado = (espaciador * int((anchoFormato / divisiones - len(item))/2)) + item + (espaciador * int((anchoFormato / divisiones - len(item))/2))
    else:
#         ToDo: Colocar un error handler para utilizar en otro contexto
        return "Alineacion incorrecta"
    return itemFormateado
    
    
# Se anida funcion principal para mantener orden
def main():
#     Se inicializan variables de formato para mejorar legibilidad de codigo
    anchoFormato = 64
    separador = ("\n" + ("-" * anchoFormato) + "\n")
    textoSalir = "Ingreso salir, imprimiendo Boleta"
    columnas = ["codigo", "producto", "cantidad", "subtotal"]
    iva = 0.19

    
#     Se inicialisa lista de productos para evitar repeticion de nombre en productos y masterdata donde guardaremos los productos
    listaproductos = []
    masterData = []
    print("¡¡¡Bienvenido al minimarket!!!", end = separador)
    
#     Generamos el loop principal para iterar los ingresos
    while True:
        print("Para salir ingrese 'salir'\nIngrese su producto, precio y cantidad que llevara")
        
#         Se solicita el producto y se valida el texto para verificar si quiere salir o ingreso es valido.
#         Esto lo repetimos para precio y cantidad, finalmente agregamos el producto a masterData
        producto = input("Ingrese producto: \n")
        if re.search("salir", producto):
            print(separador + textoSalir, end = separador)
            break
        
        while True:
            if producto in listaproductos:
                producto = input("Producto ya ingresado, intente nuevamente:\n")
                pass
            elif len(producto) <= 2 or len(producto) >=14:
                producto = input("Debe ingresar entre 2 y 14 caracteres:\n")
                pass
            elif producto.isdigit():
                producto = input("ingrese almenos 1 letra, intente nuevamente:\n")
                pass
            else:
                break
                
        
        listaproductos.append(producto)

        precio = input("Ingrese precio: \n")
        if re.search("salir", precio):
            print(separador + textoSalir, end = separador)
            break
        precio = ValidarDigito(precio)

        cantidad = input("Ingrese cantidad comprada: \n")
        if re.search("salir", cantidad):
            print(separador + textoSalir, end = separador)
            break
        cantidad = ValidarDigito(cantidad)

        subtotal = int(precio) * int(cantidad)
        codigo = GeneradorCodigo(masterData)
#         Se agregan los datos a masterData y se reinicia loop
        masterData.append([codigo, producto, precio, cantidad, subtotal])
        print("Producto ingresado correctamente.", end = separador)

#     Se evalua si se ingreso productos a la masterData
    if len(masterData) == 0:
        print("No se detecto productos ingresados\nno se puede imprimir la boleta")
    else:
        print(FormatearItem("> B O L E T A <", anchoFormato, 1, "C", "="), end = separador)
        sumSubtotal = 0
        
#         Se genera con formato columnar la boleta y se da forma con las funciones anteriores
        print(FormatearItem(columnas[0], anchoFormato, alineacion = "L"), end = "")
        print(FormatearItem(columnas[1], anchoFormato, alineacion = "C"), end = "")
        print(FormatearItem(columnas[2], anchoFormato, alineacion = "C"), end = "")
        print(FormatearItem(columnas[3], anchoFormato, alineacion = "R"), end = separador)
                    
        for item in masterData:
            item_0 = FormatearItem(FormatearCodigo(item[0]), anchoFormato)
            item_1 = FormatearItem(item[1], anchoFormato, alineacion = "C")
            item_3 = FormatearItem(item[3], anchoFormato, alineacion = "C")
            item_4 = FormatearItem(str(item[4]), anchoFormato, alineacion = "R")
            print(f"{item_0}{item_1}{item_3}{item_4}")
            sumSubtotal += item[4]
        print(" " + "-" * anchoFormato)
        
#         Finalmente se calcula el subtotal, iva y total y se muestra con formato gracias a nuestra funcion.
        descripcion = FormatearItem("Subtotal", anchoFormato, 2, espaciador = " ")
        valor = FormatearItem(str(sumSubtotal), anchoFormato, 2, "R", " ")
        print(f"{descripcion}{valor}")
        
        ivaDelSub = sumSubtotal * iva
        descripcion = FormatearItem("Iva", anchoFormato, 2, espaciador = " ")
        valor = FormatearItem(str(ivaDelSub), anchoFormato, 2, "R", " ")
        print(f"{descripcion}{valor}")
        
        total = ivaDelSub + sumSubtotal
        descripcion = FormatearItem("Total", anchoFormato, 2, espaciador = "_")
        valor = FormatearItem(str(total), anchoFormato, 2, "R", "_")
        print(f"{descripcion}{valor}", end = separador)

main()
```
